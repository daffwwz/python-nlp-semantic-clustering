```python

# import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
import re
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/144.0.0.0 Safari/537.36"
}
time.sleep(2)
```

```python
# site 1
URL = 'https://www.prospects.ac.uk/postgraduate-study/masters-degrees/postgraduate-interview-questions/'
page = requests.get(URL, headers=headers)

# parse content into bs object
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())

results = soup.find(id="main")
questions_1 = []
for h in soup.find_all("h2"): # find heading 2
    text = h.get_text(" ", strip=True) # get text inside the <div> tag
    if text.endswith("?"): # filter only questions
        questions_1.append(text) # collect data
questions_1
```

```python
# website 2
URL = 'https://goingto.university/applying-to-university/postgraduate-interview-questions/'
page = requests.get(URL, headers=headers)
soup = BeautifulSoup(page.content, "html.parser")
headings = soup.find_all("h3")
print(headings)

questions_2 = []
for h in headings:
    text = h.get_text(" ", strip=True)

    # filter only questions and initial prefixes
    if text.endswith("?") or any(text.lower().startswith(prefix) for prefix in ["why", "what", "where", "how", "tell", "have", "do", "describe"]):
        questions_2.append(text)
questions_2
```

```python
# website 6
URL = 'https://www.uc.edu/news/articles/2025/05/grad-school-interview-questions.html'
page = requests.get(URL, headers=headers)
soup = BeautifulSoup(page.content, "html.parser")
headings = soup.find_all("h3")

questions_6 = []
for h in headings:
    text = h.get_text(" ", strip=True)

    # remove leading number + dot
    text = re.sub(r'^\d+\.\s*', '', text)

    # remove quotes (smart or normal)
    text = text.strip('“”"')

    if text.endswith("?")  or any(text.lower().startswith(prefix) for prefix in [
        "why", "what", "where", "how", "tell", "have", "do", "tell"
    ]):
        questions_6.append(text)

questions_6 = questions_6[:-2]
questions_6
```
